```python
import pandas as pd
import json
import numpy as np
```

```python
LBL_FILE = '/home/lojze/data/datasets/LaRS/image_ann.json'
```

```python
with open(LBL_FILE, 'r') as file:
    data = json.load(file)
```

```python
data.keys()
```

```python
df_data = []
for ann in data['annotations']:
    d = {
        'image': ann['file_name'],
        'scene_type': ann['labels']['scene_type'],
        'lighting': ann['labels']['lighting'],
        'reflections': ann['labels']['reflections'],
        'waves': ann['labels']['waves'],
        'sp_extra_dark': ann['labels']['special']['extra_dark'],
        'sp_extra_bright': ann['labels']['special']['extra_bright'],
        'sp_glitter': ann['labels']['special']['glitter'],
        'sp_dirty_lens': ann['labels']['special']['dirty_lens'],
        'sp_wakes': ann['labels']['special']['wakes'],
        'sp_rain': ann['labels']['special']['rain'],
        'sp_fog': ann['labels']['special']['fog'],
        'sp_plants_debris': ann['labels']['special']['plants_debris'],
    }
    df_data.append(d)

df = pd.DataFrame(df_data).set_index('image')
df
```

```python
N = len(df)
N
```

```python
count = df['scene_type'].value_counts()
pd.concat([count, count / N], axis=1)
```

```python
count = df['lighting'].value_counts()
pd.concat([count, count / N], axis=1)
```

```python
count = df['reflections'].value_counts()
pd.concat([count, count / N], axis=1)
```

```python
count = df['waves'].value_counts()
pd.concat([count, count / N], axis=1)
```

```python
sp_cols = [c for c in df.columns if c.startswith('sp_')]
sp_count = df[sp_cols].sum()
pd.concat([sp_count, sp_count / N], axis=1)
```
