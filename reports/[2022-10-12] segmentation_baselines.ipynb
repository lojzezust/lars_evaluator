```python
import json
import pandas as pd
```

```python
METHODS = [
    ('BiSeNetv1', 'ResNet-50', '../results/bisenetv1_r50-d32_in1k-pre_4x4_1024x1024_160k_lars.json'),
    ('BiSeNetv2', '-', '../results/bisenetv2_fcn_4x4_1024x1024_160k_lars.json'),
    ('DeepLabv3', 'ResNet-101', '../results/deeplabv3_r101-d8_512x1024_40k_lars.json'),
    # ('DeepLabv3+', 'ResNet-50', '../results/deeplabv3plus_r50-d8_512x1024_40k_lars.json'),
    ('DeepLabv3+', 'ResNet-101', '../results/deeplabv3plus_r101-d8_512x1024_40k_lars.json'),
    ('FCN', 'ResNet-50', '../results/fcn_r50-d8_512x1024_40k_lars.json'),
    ('FCN', 'ResNet-101', '../results/fcn_r101-d8_512x1024_40k_lars.json'),
    ('UNet', 'S5', '../results/fcn_unet_s5-d16_4x4_512x1024_160k_lars.json'),
    ('PointRend', 'ResNet-101', '../results/pointrend_r101_512x1024_80k_lars.json'),
    ('SegFormer', 'MiT-B2', '../results/segformer_mit-b2_8x1_1024x1024_160k_lars.json'),
    ('Segmenter', 'ViT-B', '../results/segmenter_vit-b_mask_8x1_512x512_160k_lars.json'),
    ('STDC1', '-', '../results/stdc1_in1k-pre_512x1024_80k_lars.json'),
    ('STDC2', '-', '../results/stdc2_in1k-pre_512x1024_80k_lars.json'),
    ('KNet', 'Swin-T', '../results/knet_s3_upernet_swin-t_8x2_512x512_adamw_80k_lars.json'),
    ('WaSR (R101)', 'ResNet-101', '../results/wasr_lars_v3.json'),
    ('WaSR-T (R101)', 'ResNet-101', '../results/wasrt_lars_v3.json'),
    ('CSANet', 'ResNet-101', '../results/CSANet.json'),
    ('TMANet', 'ResNet-50', '../results/TMANet.json'),
]
```

```python
data = []
for method_name, backbone, results_path in METHODS:
    with open(results_path, 'r') as file:
        results = json.load(file)
    results['method'] = method_name
    results['backbone'] = backbone
    data.append(results)

df = pd.DataFrame(data).set_index(['method', 'backbone'])
```

```python
# df.round(2).sort_values('F1')
df.round(2)
```

```python
out_df = df[['WE_acc', 'Pr', 'Re', 'F1', 'mIoU', 'IoU_obstacle', 'IoU_water', 'IoU_sky']] * 100
out_df.round(1)
```
