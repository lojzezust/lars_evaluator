```python
import pandas as pd
import os.path as osp
import json
import unicodedata
```

```python
method_names = [
    'panoptic_deeplab_rn50_s32',
    'panoptic_fpn_r50_fpn_10x_LaRS',
    'max_deeplab_s_1025_50k',
    'mask2former_swin-b-p4-w12-384-in21k_lsj_8x2_20x_LaRS'
]

results_path = "/home/lojze/development/lars_evaluator/results_panoptic"
```

```python
dfs = [pd.read_csv(osp.join(results_path, method + '_frames.csv')) for method in method_names]
```

```python
dfs[0]
```

```python
for df in dfs:
    df['F1'] = 2 * df['TP'] / (2 * df['TP'] + df['FN'] + df['FP'])
```

```python
feature = 'aRQ_th'
df_data = {'image': dfs[0]['image']}
df_data.update({mname:df[feature] for df,mname in zip(dfs, method_names)})
df_cmp = pd.DataFrame(df_data).set_index('image')
df_cmp
```

```python
df_cmp['spread'] = df_cmp.max(axis=1) - df_cmp.min(axis=1)
df_cmp_sorted = df_cmp.sort_values('spread', ascending=False)
```

```python
df_cmp_sorted[df_cmp_sorted['spread']!=100].iloc[20:40]
```

## Image features

```python
IMAGE_ANN = '/home/lojze/data/datasets/LaRS/image_ann.json'
with open(IMAGE_ANN, 'r') as file:
    data = json.load(file)
```

```python
df_data = []
for ann in data['annotations']:
    img_name = unicodedata.normalize('NFD', ann['file_name']).encode('ascii', 'ignore').decode('utf-8') # Normalize accent characters
    d = {
        'image': osp.splitext(img_name)[0],
        'scene_type': ann['labels']['scene_type'],
        'lighting': ann['labels']['lighting'],
        'reflections': ann['labels']['reflections'],
        'waves': ann['labels']['waves'],
        'sp_extra_dark': ann['labels']['special']['extra_dark'],
        'sp_extra_bright': ann['labels']['special']['extra_bright'],
        'sp_glitter': ann['labels']['special']['glitter'],
        'sp_dirty_lens': ann['labels']['special']['dirty_lens'],
        'sp_wakes': ann['labels']['special']['wakes'],
        'sp_rain': ann['labels']['special']['rain'],
        'sp_fog': ann['labels']['special']['fog'],
        'sp_plants_debris': ann['labels']['special']['plants_debris'],
    }
    df_data.append(d)

df_cls = pd.DataFrame(df_data).set_index('image')
df_cls
```

```python
df_cmp_p = df_cmp.join(df_cls, how='left')
```

```python
df_cmp_sorted = df_cmp_p[df_cmp_p['sp_plants_debris'] == True].sort_values('spread', ascending=False)
```
